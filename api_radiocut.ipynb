```python
import requests
import time

def fetch_url(url):
    response = requests.get(url)
    return response

urls = ["https://radiocut.fm/api/search/?type=cut&page=160494",
 "https://radiocut.fm/api/search/?type=cut&page=160495"]

crawl_delay = 2  # segundos

docs = []
for url in urls:
    response = fetch_url(url)
    if response.status_code == 200:
        print("Fetched:", url)
        docs.append(response.text)

    else:
        print("Failed to fetch:", url)
    
    # Esperar de acuerdo al tiempo del crawl (rebaje)
    time.sleep(crawl_delay)
```
